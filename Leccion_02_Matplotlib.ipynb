## *️⃣ Funciones básicas de configuración
En matplotlib debemos llamar a varias funciones para hacer una sola gráfica. Por eso antes de ver las funciones para hacer gráficas vamos a ver cuáles son las funciones básicas de configuración.

#### *️⃣ Figure

```python
# El gráfico tendrá tamaño 2 de ancho x 3 de alto y tendrá color de fondo azul claro
plt.figure(figsize=(2, 3), facecolor='lightblue')
```

Esta función configura los parámetros básicos del gráfico que definiremos a continuación, por lo tanto **¡si queremos usar esta función debemos llamarla antes de cualquier otra!**

No es obligatorio llamar a esta función. Si lo lo hacemos matplotlib usará las configuraciones por defecto.




La sintaxis es:

 **plt.figure(** *configuraciones* **)**

##### ——— 𝒄𝒐𝒏𝒇𝒊𝒈𝒖𝒓𝒂𝒄𝒊𝒐𝒏𝒆𝒔: ———

**figsize**

*Tamaño de todo el gráfico.*

Se le pasa una tupla con el tamaño deseado en pulgadas.

**facecolor**

*Color de fondo del gráfico*

Puede ser un RGB en hexadecimal o puede ser un string con un nombre de color en formato X11. Algunos ejemplos:

* 'red'
* 'darkred'
* 'purple'
* 'mediumpurple'

Lista de colores en formato X11:

https://en.wikipedia.org/wiki/X11_color_names

**edgecolor**

*Color del borde del gráfico.*

Se le pasa de la misma forma que en facecolor.

**linewidth**

*Ancho del borde del gráfico.*

Se le pasa un número.

## Gráficos de dispersión

**plt.scatter(valores_x, valores_y,** *configuraciones* **)**

*Nombre que se usará en la leyenda para ese gráfico.*

*Especifica el marcador para cada punto en la línea.*

* 'x' para una x


**c**

*Define los colores de los puntos.*

Puedes pasarle:

* Un array de **valores numéricos**, en cuyo caso se asignarán colores automáticamente utilizando una escala de colores (por ejemplo, colores más hacia el rojo para los valores más bajos y colores más hacia el verde para los valores más altos)

**s**

*Define el tamaño de los puntos.*

Puedes pasarle:

* Un solo valor que se aplicará a todos los puntos

##### **Ejemplo**

```python
# Crear datos de dispersión
import numpy as np
rng = np.random.default_rng()
x_scatter = rng.random(50)
y_scatter = rng.random(50)
print(x_scatter)

# Crear un gráfico de dispersión
plt.figure(figsize=(8, 5))
plt.scatter(x_scatter, y_scatter, c=y_scatter, label='Datos de dispersión')

# Personalizar el gráfico
plt.title('Gráfico de Dispersión')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Mostrar el gráfico
plt.show()
```

## Gráfico de torta

**plt.pie(sizes, labels= *tus etiquetas*,** *más configuraciones* **)**

##### ——— 𝒄𝒐𝒏𝒇𝒊𝒈𝒖𝒓𝒂𝒄𝒊𝒐𝒏𝒆𝒔: ———

**labels**

*Etiquetas para cada porción del pastel.*

Le pasas una lista.


**colors**

*Colores para cada porción del pastel.*

Le pasas una lista de nombres de colores o códigos hexadecimales.

Si no agregas este argumento matplotlib usa colores por defecto.


**autopct**

*Formato de los porcentajes.*

Le pasas una cadena de formato para especificar cómo se deben etiquetar los porcentajes. Por ejemplo, '%1.1f%%' mostrará cada número con un solo decimal y al final un signo de porcentaje.

Si no agregas este argumento no se mostrarán los porcentajes de cada porción.


**startangle**

*Ángulo en el que empieza a graficar.*

Se le pasa un número.

La gráfica se hace en sentido antihorario. El ángulo de inicio por defecto es 0 grados (el lado derecho del círculo).

**explode**

*Permite sacar una porción hacia afuera más que las otras.*

Se le pasa una secuencia de valores que especifica el desplazamiento de una porción del pastel desde el centro. Si un valor es 0, la porción no está desplazada. Si el valor es mayor que 0, la porción se desplazará hacia afuera.

##### **Ejemplo**

```python
# Datos para el gráfico de torta
porciones = [25, 30, 15, 10, 20]
etiquetas = ['Manzanas', 'Bananos', 'Uvas', 'Naranjas', 'Fresas']
colores = ['red', 'yellow', 'mediumpurple', 'orange', 'pink']
sobresalir = (0.15, 0, 0, 0, 0)

# Crear el gráfico de torta con opciones de estilo
plt.pie(porciones, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=90, explode=sobresalir)

# Añadir título
plt.title('Gráfico de Torta')

# Mostrar el gráfico
plt.show()
```

##### **Ejemplo de NLP: Análisis de sentimientos**

Ahora utilizaremos Matplotlib para realizar un análisis de sentimientos simple en un conjunto de comentarios de usuarios. Crearemos un gráfico de pastel para visualizar la proporción de comentarios positivos, negativos y neutros.

```python
# Comentarios de usuarios
comentarios = [
    'Me encanta este producto, es increíble',
    'No estoy seguro, la calidad podría ser mejor',
    'El servicio al cliente es muy amable y útil',
    'No recomendaría este producto a otros',
    'Buena relación calidad-precio',
    'No tengo ninguna queja, todo está bien'
]

# Etiquetas de sentimiento (positivo, negativo, neutro)
sentimientos = ['positivo', 'negativo', 'positivo', 'negativo', 'positivo', 'neutro']
```

```python
# Contar la frecuencia de cada etiqueta de sentimiento
frecuencia_sentimientos = dict(zip(*np.unique(sentimientos, return_counts=True)))

# Colores para el gráfico de pastel
colores = ['lightgreen', 'lightcoral', 'lightskyblue']

# Crear el gráfico de pastel
plt.pie(frecuencia_sentimientos.values(), labels=frecuencia_sentimientos.keys(), colors=colores, autopct='%1.1f%%')

# Añadir título
plt.title('Análisis de Sentimientos en Comentarios de Usuarios')

# Mostrar el gráfico
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear una figura y una matriz de subgráficos 2x2
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Subgráfico 1
axs[0, 0].plot(x, y, color='hotpink')

# Subgráfico 2
axs[0, 1].plot(x, y, color='deepskyblue')

# Subgráfico 3
axs[1, 0].plot(x, y, color='limegreen')

# Subgráfico 4
axs[1, 1].plot(x, y, color='orchid')

# Mostrar el gráfico
plt.show()
```

**𝒄𝒐𝒏𝒇𝒊𝒈𝒖𝒓𝒂𝒄𝒊𝒐𝒏𝒆𝒔:** las mismas que para las líneas de cuadrícula normales de matplotlib

##### **Ejemplo**

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear una figura y una matriz de subgráficos 2x2
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
fig.suptitle('Un título para gobernarlos a todos', fontsize='xx-large')

# Subgráfico 1
axs[0, 0].plot(x, y, color='hotpink', label='Siiiii')
axs[0, 0].set_title('Hola1')
axs[0, 0].legend()

# Subgráfico 2
axs[0, 1].plot(x, y, color='deepskyblue')
axs[0, 1].set_title('Hola2')
axs[0, 1].grid()

# Subgráfico 3
axs[1, 0].plot(x, y, color='limegreen')
axs[1, 0].set_title('Hola3')

# Subgráfico 4
axs[1, 1].plot(x, y, color='orchid')
axs[1, 1].set_title('Hola4')
axs[1, 1].set_xlabel('Etiqueta X')
axs[1, 1].set_ylabel('Etiqueta Y')

# Mostrar el gráfico
plt.show()
```

#### ✴️ Ajuste automático


**plt.tight_layout()**

Es una función que ajusta automáticamente los espaciados entre subgráficos o entre subgráficos y los bordes de la figura.

##### **Ejemplo**

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear una figura y una matriz de subgráficos 2x2
fig, axs = plt.subplots(2, 2, figsize=(8, 6))
fig.suptitle('Un título para gobernarlos a todos', fontsize='xx-large')

# Subgráfico 1
axs[0, 0].plot(x, y, color='hotpink', label='Siiiii')
axs[0, 0].set_title('Hola1')
axs[0, 0].legend()

# Subgráfico 2
axs[0, 1].plot(x, y, color='deepskyblue')
axs[0, 1].set_title('Hola2')
axs[0, 1].grid()

# Subgráfico 3
axs[1, 0].plot(x, y, color='limegreen')
axs[1, 0].set_title('Hola3')

# Subgráfico 4
axs[1, 1].plot(x, y, color='orchid')
axs[1, 1].set_title('Hola4')
axs[1, 1].set_xlabel('Etiqueta X')
axs[1, 1].set_ylabel('Etiqueta Y')

# AJUSTE AUTOMÁTICO
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# FRECUENCIA PALABRAS

# Calculamos la frecuencia de cada palabra manualmente (se podría hacer con la librería Counter)
# y la guardamos en un diccionario
frecuencia_palabras = {}
for palabra in palabras:
    if palabra in frecuencia_palabras:
        frecuencia_palabras[palabra] += 1
    else:
        frecuencia_palabras[palabra] = 1
print('frecuencia_palabras: ', frecuencia_palabras)

# Hacemos una lista con todas las frecuencias
solo_frecuencias = list(frecuencia_palabras.values())
print('solo_frecuencias: ', solo_frecuencias, '\n')

# Ordenamos las palabras por frecuencia en una lista
palabras_ordenadas_por_frecuencia = sorted(frecuencia_palabras, key=frecuencia_palabras.get, reverse=True)

# Ponemos las 5 palabras más comunes en una lista y sus frecuencias en otra lista
palabras_mas_comunes = palabras_ordenadas_por_frecuencia[0:5]
frecuencias_palabras_mas_comunes = [frecuencia_palabras[palabra] for palabra in palabras_mas_comunes]

print('palabras_mas_comunes: ', palabras_mas_comunes)
print('frecuencias_palabras_mas_comunes: ', frecuencias_palabras_mas_comunes)
```

```python
# LONGITUD PALABRAS

# Hacemos una lista con la longitud de cada palabra
longitudes = [len(palabra) for palabra in palabras]
print('longitudes: ', longitudes)

# Ordenamos las palabras por longitud en una lista
palabras_ordenadas_por_longitud = sorted(palabras, key=len, reverse=True)

# Ponemos las 5 palabras más largas en una lista y sus longitudes en otra lista
palabras_mas_largas = palabras_ordenadas_por_longitud[0:3]
longitud_palabras_mas_largas = [len(palabra) for palabra in palabras_mas_largas]

print('palabras_mas_largas: ', palabras_mas_largas)
print('longitud_palabras_mas_largas: ', longitud_palabras_mas_largas)
```
