**Array creado con un rango**

Hay dos formas de hacer esto:
* **np.arange** permite especificar el tamaño del paso
* **np.linspace** permite especificar la cantidad de puntos que deseas en el array

#### np.arange

Sintaxis: **np.arange(comienzo, final, tamaño del paso)**

* ¡El número final no se incluye, siempre llega hasta el número anterior! (Si quieres que llegue hasta 8 debes poner 9 como número final)
* El comienzo por defecto es 0 y el tamaño de paso por defecto es 1.

```python
# Definamos un array que vaya de 2 a 7:
a = np.arange(2, 8)
print(a)
```

```python
# Definamos un array que vaya de 0 a 9, de 3 en 3
b = np.arange(0, 10, 3)
print(b)
```

#### np.linspace

Sintaxis: **np.linspace(comienzo, final, cantidad de pasos)**

Se incluye el valor del comienzo y también el del final.

```python
# Definamos un array que vaya de 0 a 10, con 5 pasos
c = np.linspace(0, 11, 5)
print(c)
```

### Manipulación de un array
*Estas operaciones se aplican a un array y devuelven un array.*


**Reshape**

```python
# Crear un array de una dimensión con 12 elementos (de 0 a 12)
array_original = np.arange(0, 12)
print("Array original:")
print(array_original)

# Redimensionar el array para que tenga 3 filas y 4 columnas
array_redimensionado = array_original.reshape(3, 4)
print("\nArray redimensionado:")
print(array_redimensionado)
```

**Transpose**

```python
mi_array = np.array([[1, 2, 3], [4, 5, 6]])
print("Array original:")
print(mi_array)

transpuesto = mi_array.transpose()
print("\nArray transpuesto:")
print(transpuesto)
```

**Operación especial: .dot**

Corresponde a la operación Producto Matricial si estamos trabajando con arrays de más de una dimensión y corresponde al Producto Punto de dos arrays si ambos son de una dimensión.


```python
# Crear dos matrices
matriz1 = np.array([[1, 2, 3],
                    [4, 4, 4]])
print("Matriz 1:")
print(matriz1)

matriz2 = np.array([[2, 3],
                    [4, 0],
                    [1, 2]])

print("\nMatriz 2:")
print(matriz2)
```

```python
# Calcular el producto matricial
producto_matricial = np.dot(matriz1, matriz2)

print("Producto matricial:")
print(producto_matricial)
```

### Generación de números aleatorios
Siempre que vayamos a generar números aleatorios debemos crear un generador.

* Un "generador" es un objeto instanciado de la clase **np.random.Generator**

* La forma recomendada de crear un generador es llamando a la función **np.random.default_rng()**, esa función nos devuelve un objeto de tipo generador y ese lo guardamos en la variable **rng** por convención

```python
# Debemos escribir esto siempre antes de crear números aleatorios:
rng = np.random.default_rng()

# Vemos  que dentro de "rng" quedó un generador:
print(rng)
```

**rng.random**

Función dentro de nuestro generador que sirve para generar números aleatorios entre 0 y 1 de tipo float

```python
rng = np.random.default_rng()

# Si llamas la función sin pasarle nada te devuelve un sólo número aleatorio
random_number = rng.random()
print(random_number)
```

```python
rng = np.random.default_rng()

# Si llamas la función pasándole un shape te devuelve un
# array con esa shape lleno de números aleatorios entre 0 y 1
random_array = rng.random((3, 2))
print(random_array)
```

**rng.integers**

Función dentro de nuestro generador que sirve para generar números aleatorios de tipo int entre dos números que le digamos

Sintaxis: **rng.integers(*inicio*, *fin*, size= *shape que queramos*)**
* inicio si se incluye
* final no se incluye (va hasta el anterior)

```python
rng = np.random.default_rng()

# Si no le pasas shape te devuelve un solo número
random_integer = rng.integers(1, 10)
print(random_integer)
```

```python
rng = np.random.default_rng()

# Si le pasas un shape te devuelve un array
random_int_array = rng.integers(1, 10, size=(3, 2))
print(random_int_array)
```

## Cheat-Sheet Numpy

### Funciones matemáticas:

*Se pueden aplicar a un elemento o a un array.*

*Si se aplican a un elemento devuelven un elemento, si se aplican a un array devuelven un array.*

* **np.sin, np.cos, np.tan:** Funciones trigonométricas.
* **np.sqrt:** Raíz cuadrada.
* **np.power:** Potenciación a cualquier exponente.

### Creación de arrays:
* **np.array:** Crea un array a partir de una lista o tupla.
* **np.zeros:** Crea un array de ceros.
* **np.ones:** Crea un array de unos.
* **np.identity:** Crea un array identidad.
* **np.arange:** Crea un array con valores espaciados uniformemente dentro de un rango.
* **np.linspace:** Crea un array con valores espaciados uniformemente entre dos números.

### Operaciones sobre un array que devuelven un valor:
* **np.sum:** Devuelve la suma de todos los valores en del array.
* **np.max:** Devulve en valor máximo en el array.
* **np.min:** Devulve en valor mínimo en el array.
* **np.average:** Devuelve el promedio de todos los valores en el array.

### Manipulación de un array:
*Estas operaciones se aplican a un array y devuelven un array.*
* **np.reshape:** Cambia la forma de un array.
* **np.transpose:** Transpone un array.

### Operaciones entre arrays:

*Estas operaciones se aplican a 2 arrays y devuelven un array.*

* **arr1 + arr2:** Suma dos arrays elemento por elemento.
* **arr1 - arr2:** Resta dos arrays elemento por elemento.
* **arr1 * arr2:** Multiplica dos arrays elemento por elemento.
* **arr1 / arr2:** Divide dos arrays elemento por elemento.

Operación especial:

*Esta operación se aplica a 2 arrays y devuelve un valor.*
* **np.dot:** Corresponde a la operación Producto Matricial si estamos trabajando con arrays de más de una dimensión y corresponde al Producto Punto de dos arrays si ambos son de una dimensión.

***rng = np.random.default_rng()***

*Dentro de rng quedó guardado nuestro generador.*
