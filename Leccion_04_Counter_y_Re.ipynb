## **Counter**

El módulo **Counter** de Python es una parte de la librería **collections** y se utiliza para contar la ocurrencia de elementos en una secuencia, como una lista o una cadena de texto.

````python
# Sintaxis para importar Counter
from collections import Counter
````

#### **Crear un objeto Counter**

Puedes crear un objeto **Counter** pasando una secuencia como argumento. Este objeto contendrá las frecuencias de los elementos en la secuencia.

**Para contar elementos en una cadena de texto**

````python
string_ejemplo = 'Abracadabra'

# Usar Counter para contar los caracteres
mi_conteo = Counter(string_ejemplo)
print(mi_conteo)
````

**Para contar elementos en una lista**

````python
lista_ejemplo = ['lirio', 'rosa', 'tulipán', 'margarita', 'rosa', 'tulipán', 'lirio', 'tulipán', 'rosa', 'tulipán', 'margarita', 'margarita', 'rosa', 'rosa']

# Usar Counter para contar elementos
conteo = Counter(lista_ejemplo)
conteo
````

**Para contar elementos en una lista de listas**


````python
lista_de_listas = [['hola', 'mi', 'nombre', 'es', 'pedro'], ['hola', 'mi', 'nombre', 'es', 'camila']]

# Transformar la lista en una lista plana (usando comprensión de listas)
lista_plana = [elemento for sublista in lista_de_listas for elemento in sublista]
print('lista_plana: ', lista_plana)

# Usar Counter para contar los elementos repetidos
contador = Counter(lista_plana)
print(contador)
````

#### **Funciones de Counter**

##### **.items()**

Devuelve un **objeto de vista** en el que hay tuplas de cada elemento y su conteo.

* *Un objeto de vista en Python es un tipo de objeto que proporciona una vista o representación de los datos almacenados en otro objeto (como una lista, diccionario, etc) sin copiar esos datos.*

* Si lo ponemos dentro de la función **list()** nos devolverá una **lista** con tuplas de cada elemento y su conteo.

````python
# Así sin más
vista_elementos_y_conteo = conteo.items()
vista_elementos_y_conteo
````

````python
# Convirtiéndolo en una lista
lista_elementos_y_conteo = list(conteo.items())
lista_elementos_y_conteo
````

##### **.elements()**

Devuelve un **iterador** que produce todos los elementos del objeto Counter en el orden en que están ahí.

* *Un iterador en Python es un objeto que permite recorrer un contenedor (como una lista, tupla, diccionario, etc) y acceder a sus elementos uno por uno.*

* Si lo ponemos dentro de la función **list()** nos devolverá una **lista** con todos los elementos del objeto Counter.

````python
# Creamos el objeto iterador de los elementos
iterador_elementos = conteo.elements()
iterador_elementos
````

````python
# Así se usaría el objeto iterador
for elemento in iterador_elementos:
    print(elemento)
````

````python
# Así convertimos el objeto iterador en una lista
lista_elementos = list(conteo.elements())
lista_elementos
````

##### **.most_common(n)**

Devuelve una lista de tuplas con los n elementos más comunes y sus conteos.

````python
# Obtener los 3 elementos más comunes
elementos_comunes = conteo.most_common(3)
elementos_comunes
````

##### **.update()**

Actualiza el conteo, es decir, cuenta elementos nuevos que le pasemos junto con lo que ya había contado.

````python
nuevos_elementos = ['margarita', 'margarita', 'margarita', 'margarita', 'clavel', 'clavel', 'girasol']

# Actualizar el conteo
conteo.update(nuevos_elementos)
conteo
````

##### **.substract()**

Resta elementos de otro objeto Counter o de cualquier secuencia.

````python
elementos_a_restar = ['girasol', 'clavel']

# Restar el conteo
conteo.subtract(elementos_a_restar)
conteo
````

##### **.total()**

Devuelve la suma de todos los conteos.

````python
total = conteo.total()
total
````

##### **.clear()**

Reinicia el objeto tipo Counter.

````python
conteo.clear()
conteo
````

#### `[] = conjunto`

Podemos usar **corchetes cuadrados** para declarar **conjuntos válidos de caracteres.**

Para explicar en este notebook usaré este formato:


```
patrón = qué cosa se considera que cumple con ese patrón
```



Dentro de los corchetes cuadrados podemos usar un guión para hacer un intervalo de letras o números:

```
[a-z] = un caracter de la 'a' a la 'z' (una letra en minúsculas)

[A-Z] = un caracter de la 'A' a la 'Z' (una letra en mayúsculas)

[a-d] = un caracter de la 'a' a la 'd' ('a', 'b', 'c' o 'd')

[0-9] = un número del 0 al 9

[3-6] = un número del 3 al 6 (3, 4, 5 o 6)

```

Poner algunos caracteres sin guión dentro de unos corchetes cuadrados significa que puede ser cualquiera de esos caracteres:

```
[anfu] = un caracter que sea la 'a', la 'n', la 'f' o la 'u'

[#_!] = un caracter que sea '#', '_' o '!'

```







Poner **varios intervalos distintos** dentro de unos mismos corchetes significa que puede ser un caracter de cualquiera de esos intervalos. No hay que separar los intervalos por comas ni por espacios:


```
[a-zA-Z] = un caracter de la 'a' a la 'z' o de la 'A' a la 'Z'
(cualquier letra en minúsculas o mayúsculas)

[a-z0-9] = un caracter de la 'a' a la 'z' o un número del 0 al 9
(cualquier letra en minúsculas o cualquier número)

[b-d5-7@] = un caracter de la 'b' a la 'd', o un número del 5 al 7
o el caracter @ ('b', 'c', 'd', 5, 6, 7 o @)
```

**Nota:** cuando escribimos a-z o A-Z **no está incluida la letra ñ**. Si queremos incluir la ñ debemos agregarla por aparte:


```
[a-zñ] = un caracter de la 'a' a la 'z', incluyendo la 'ñ' (una letra en minúsculas)

[A-ZÑ] = un caracter de la 'A' a la 'Z', incluyendo la 'Ñ' (una letra en mayúsculas)
```

Poner un **símbolo + justo después de los corchetes** significa que no estamos buscando sólo un caracter que pertenezca a ese intervalo sino **uno o más caracteres** que pertenezcan a ese intervalo:



```
[a-z] = un caracter de la 'a' a la 'z' (una letra en minúsculas)

[a-z]+ = uno o más caracteres de la 'a' a la 'z' (una o más letras en minúsculas)

```



### 🛅 **Funciones básicas**

#### `re.search()`

Buscará si la cadena cumple con el patrón en algún punto (no necesariamente al comienzo de la cadena).

* Devuelve un objeto de tipo match si encuentra una coincidencia
* Devuelve None si no encuentra ninguna coincidencia.


````python
# Definir el patrón
patron = "natural"

# Definir la cadena en la que queremos buscar ese patrón
cadena = "Procesamiento de lenguaje natural"

# Usar re.search() para buscar el patrón en la cadena
resultado = re.search(patron, cadena)
print(resultado)
````

El objeto tipo match que nos devuelve re.search() tiene algunas funciones que podemos usar:

* **resultado.start():** Devuelve la posición de inicio de la coincidencia en la cadena.
* **resultado.end():** Devuelve la posición de fin de la coincidencia en la cadena.
* **resultado.group():** Devuelve la cadena coincidente.

````python
# Definir el patrón
patron = "gato"

# Definir la cadena en la que queremos buscar ese patrón
cadena = "Mi gato es azul."

# Usar re.search() para buscar el patrón en la cadena
resultado = re.search(patron, cadena)
print(resultado, '\n')

if resultado:
    print("Se encontró una coincidencia.")
    print("Inicio de la coincidencia:", resultado.start())
    print("Fin de la coincidencia:", resultado.end())
    print("Texto de la coincidencia:", resultado.group())
else:
    print("No se encontró ninguna coincidencia.")
````

#### `\letra = grupo predefinido`

Un backslash seguido de una letra representa un **grupo predefinido** de caracteres.




**Dígitos**

```
\d = cualquier dígito
\D = cualquier carácter que no sea un dígito
```

**Caracteres de espacio en blanco**

```
\s = cualquier carácter de espacio en blanco
\S = cualquier carácter que no sea de espacio en blanco
```

**Caracteres alfanuméricos**

```
\w = cualquier carácter alfanumérico
\W = cualquier carácter que no sea alfanumérico
```

Los caracteres alfanuméricos son cualquier letra en minúsculas o

mayúsculas (**incluyendo tildes y acentos y la ñ**) o cualquier **dígito** o un **guión bajo**.

\



Después de cualquiera de los patrones anteriores podemos poner un símbolo + para indicar que buscamos uno o más caracteres de ese tipo.

\

### 🔑 **Patrones intermedios**

#### `^ = inicio`

- Si empezamos nuestro patrón con un circunflejo (^) intentará hacer coincidir el patrón desde **el inicio de la cadena** en la que estemos buscando.

- Es lo mismo que usar la función **.match()**




````python
# Buscamos que la cadena inicie con uno o más dígitos
patron = "^\d+"
````

````python
cadena = "solecito2246"
resultado = re.search(patron, cadena)

if resultado:
    print("Se encontró una coincidencia.")
    print("Texto de la coincidencia:", resultado.group())
else:
    print("No se encontró ninguna coincidencia.")
````

#### `$ = fin`

- Si terminamos nuestro patrón con un $ intentará hacer coincidir el patrón con **el final de la cadena** en la que estemos buscando.

#### `letra r antes del patrón`



**Nota:** de ahora en adelante cada vez que definamos una cadena como patrón pondremos la **letra r** antes de la cadena.

*Ejemplo:* patron = r"[a-z]"

\

Esto es para asegurarnos de que Python tomará la cadena como una cadena literal y **no intentará interpretar** ninguna parte de la cadena como un **caracter de escape**.

- Un **caracter de escape** en Python es un carácter especial que se utiliza para representar secuencias especiales dentro de cadenas de texto. Estas secuencias especiales se componen de una barra invertida (\) seguida de otro carácter.

\

Algunos caracteres de escape de Python



```
\n: representa un salto de línea.
\t: representa una tabulación.
\\: representa una barra invertida.
\': representa una comilla simple.
\": representa una comilla doble.
```



\

**No queremos que Python use sus propios caracteres de escape**, ya que podría interferir con los caracteres de escape de las expresiones regulares o con algunos patrones.

Por eso siempre debemos usar la sintaxis **r"patron"** cada vez que definimos un patrón.

\

También si la cadena de texto en la que estamos buscando  tiene **backslashs** (\), debemos definirla con **r"cadena"**.

\

#### `() = grupos de captura`





 Los grupos de captura en la biblioteca re de Python te permiten **etiquetar partes específicas** de una expresión regular y luego **acceder a ellas individualmente**.

\

Por ejemplo, veamos la expresión regular para extraer números de teléfono en un formato específico:

````python
patron = r"(\d+)-(\d+)-(\d+)"

texto = "Mi número de teléfono es 123-456-7890."
resultado = re.search(patron, texto)

if resultado:
    print("Número de teléfono completo:", resultado.group(0))
    print("Código del operador:", resultado.group(1))
    print("Prefijo:", resultado.group(2))
    print("Sufijo:", resultado.group(3))
else:
    print("No se encontró ningún número de teléfono.")
````

Como vemos:

- **resultado.group(0):** Devuelve la cadena completa que coincide con el patrón.

- **resultado.group(1), resultado.group(2), resultado.group(3):** Devuelven cada grupo de captura por separado.

\

También podemos hacer grupos de captura con conjuntos de caracteres (corchetes cuadrados) dentro:

````python
patron = r"([a-z1-3]+)\\([0-9-]+)"

texto = r"hola123\278-34-3"
resultado = re.search(patron, texto)

if resultado:
    print(resultado.group(0))
    print(resultado.group(1))
    print(resultado.group(2))
else:
    print("No se encontró ninguna coincidencia")
````

### 🔑 **Más patrones avanzados**

#### `cuantificadores`



Los cuantificadores son símbolos que te permiten especificar **cuántas veces un elemento puede aparecer** en una cadena de texto.

\

`* (Cero o más veces)`

Este cuantificador indica que el elemento precedente puede aparecer cero o más veces en la cadena.

Por ejemplo: **a*** coincidirá con "x", "a", "aa", "aaa", etc.

\

`+ (Una o más veces)`

Este cuantificador indica que el elemento precedente debe aparecer al menos una vez en la cadena, pero puede aparecer más veces.

Por ejemplo: **a+** coincidirá con "a", "aa", "aaa", etc., pero no con "x".

\

`? (Cero o una vez)`

Este cuantificador indica que el elemento precedente puede aparecer cero o una vez en la cadena. Esencialmente, hace que el elemento sea opcional.

Por ejemplo: **colou?r** coincidirá tanto con "color" como con "colour".

\

`{n} (Exactamente n veces)`

Este cuantificador te permite especificar exactamente cuántas veces debe aparecer el elemento precedente en la cadena.

Por ejemplo: **a{3}** coincidirá solo con "aaa".

\

`{n,} (Al menos n veces)`

Este cuantificador indica que el elemento precedente debe aparecer al menos n veces en la cadena, pero puede aparecer más veces.

Por ejemplo: **a{2,}** coincidirá con "aa", "aaa", "aaaa", etc., pero no con "a" sola.

\

`{n,m} (De n a m veces)`

Este cuantificador te permite especificar un rango de veces que el elemento precedente puede aparecer en la cadena, desde n hasta m veces, ambos incluidos.

Por ejemplo: **a{2,4}** coincidirá con "aa", "aaa" y "aaaa", pero no con "a" sola ni con "aaaaa".

````python
# Ejemplo números de teléfono

patron = r"\d{3}-\d{3}-\d{4}"

texto = "Mi número de teléfono es 123-456-7890."
resultado = re.search(patron, texto)

# Si hay coincidencia, imprimimos el número de teléfono completo y cada grupo de captura
if resultado:
    print("Número de teléfono:", resultado.group())
else:
    print("No se encontró ningún número de teléfono.")
````

````python
# Ejemplo usando todos los cuantificadores

# Patrones de ejemplo
patron_asterisco = r'a*'
patron_mas = r'a+'
patron_interrogacion =r'colou?r'
patron_exacto = r'a{3}'
patron_al_menos = r'a{2,}'
patron_rango = r'a{2,4}'

# Cadena de ejemplo
cadena = "aaa color colour"

# Coincidencias
print("Patrón con *:", re.findall(patron_asterisco, cadena))
print("Patrón con +:", re.findall(patron_mas, cadena))
print("Patrón con ?: ", re.findall(patron_interrogacion, cadena))
print("Patrón exacto:", re.findall(patron_exacto, cadena))
print("Patrón al menos:", re.findall(patron_al_menos, cadena))
print("Patrón con rango:", re.findall(patron_rango, cadena))
````

### 🚩 **Flags**

Los flags son **argumentos opcionales** que se pueden pasar a las funciones de la librería re para hacer cosas especiales.

\

Recordemos la sintaxis para trabajar con expresiones regulares en Python:

**re.función(“patrón de búsqueda”, “cadena en la que quiero buscar”, flags)**

* *Ejemplo: re.search("[a-z]", “Hola, soy una cadena”, re.I)*

\

#### Principales flags

- **re.IGNORECASE o re.I:** Busca coincidencias sin distinguir entre mayúsculas y minúsculas.

- **re.MULTILINE o re.M:** Permite que los metacaracteres ^ y $ coincidan con el comienzo y el final de cada línea, en lugar de solo el principio y el final de la cadena.

- **re.DOTALL o re.S:** Hace que el metacaracter . coincida con cualquier carácter, incluidos los saltos de línea (\n).

- **re.VERBOSE o re.X:** Permite escribir expresiones regulares de forma más legible al ignorar los espacios en blanco y los comentarios dentro de la expresión regular.

````python
# Ejemplo Multiline

texto = """Inicio: Primera línea
Algo de texto aquí
Inicio: Segunda línea
Otra línea sin Inicio al principio"""

patron = re.compile(r'^Inicio:.*', re.MULTILINE)

coincidencias = patron.findall(texto)
print(coincidencias)
````

````python
# Ejemplo Verbose

texto = "Mi número de teléfono es 123-456-7890"

patron = re.compile(r'''
                    (\d{3})  # Captura el código del operador
                    -         # Separador
                    (\d{3})  # Captura los primeros tres dígitos
                    -         # Separador
                    (\d{4})  # Captura los últimos cuatro dígitos
                    ''', re.VERBOSE)

coincidencias = patron.search(texto)

if coincidencias:
    print("Número de teléfono encontrado:", coincidencias.group())
else:
    print("Número de teléfono no encontrado")
````
