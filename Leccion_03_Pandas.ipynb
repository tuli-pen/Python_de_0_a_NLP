## Estructuras de datos en Pandas

Pandas ofrece dos estructuras de datos principales: `Series` y `DataFrame`.

### Series

Una Serie es una estructura unidimensional que puede contener cualquier tipo de datos.

Es como una columna en una hoja de cálculo.

**s = pd.Series(datos,** index=índice)
* **datos** puede ser una lista, un diccionario, un único elemento o un array de numpy
* **índice** es una lista que se usará para etiquetar el índice de cada dato. Pueden haber índices duplicados.

**Usando listas**

```python
# Crear una Serie
serie = pd.Series([10, 20, 30, 40])
print(serie)
```

```python
# Crear una Serie con un índice personalizado
serie = pd.Series([10, 20, 30, 40], index=range(8, 12))
print(serie)
```

```python
# Crear una Serie con otros índices personalizados
serie = pd.Series([10, ['abcdef', 'g'], ('si', 'no'), 'Holaaa', 'Holaaa'], index=[[1,2], (3,4), 'Holaaa', 82, 'Holaaa'])
print(serie)
```

**Usando diccionarios**

```python
# Usará las claves como índices
# Creará la serie en el orden que esté en el diccionario
dicc = {"b": 1, "a": 0, "c": 2}
mi_serie = pd.Series(dicc)
print(mi_serie)
```

```python
# Si le pasamos un diccionario y unos índices aparte buscará esos índices en el diccionario
# (NaN signfica Not a Number y es como Pandas marca un dato faltante)

dicc = {"a": 0.0, "b": 1.0, "c": 2.0}
serie = pd.Series(dicc)
print(serie, '\n')

serie_2 = pd.Series(dicc, index=["b", "c", "d", "a"])
print(serie_2)
```

**Usando un único elemento**

```python
a = pd.Series(5.934853)
print(a)
```

```python
# Cuando usamos un elemento, pero le pasamos índices, crea una fila para cada índice
a = pd.Series('Hola', index=["a", "b", "c", "d", "e"])
print(a)
```

**Usando un array de numpy**

```python
import numpy as np
rng = np.random.default_rng()
nums_random = rng.integers(10, 21, 5)
print(nums_random, '\n')

serie = pd.Series(nums_random, index=["a", "b", "c", "d", "e"])
print(serie)
```

Toda Serie tiene un nombre.

Para cambiarle el nombre usamos esta sintaxis:

s = pd.Series(datos, index=índice, **name='nombre'**)

```python
s = pd.Series('Elemento', index=[10,11,12], name="Ernesto Pérez")
print(s)
```

### DataFrame

Un DataFrame es una estructura bidimensional.

Su funcionamiento es como una **hoja de cálculo** y su estructura es como **un diccionario de Series.**

Esta es su sintaxis:

**df = pd.DataFrame(datos,** index= índices filas, columns= índices columnas )
* **datos** puede ser un array de numpy de 2 dimensiones, un diccionario, listas, series u otro DataFrame
* Se suele llamar df por convención

**Usando un array de numpy**

```python
import numpy as np
mi_array = np.array([[200,300,400], [500,600,700]])
mi_df = pd.DataFrame(mi_array)

# Nota: colab puede imprimir los DataFrames de una forma especial
print(mi_df)
```

```python
mi_array = np.array([[200,300,400], [500,600,700]])
mi_df = pd.DataFrame(mi_array, index=['uno', 'dos'], columns=[5,6,7])
mi_df
```

**Usando un diccionario**

```python
# Crear un DataFrame a partir de un diccionario de listas
# Tomará las claves como columnas
data = {'Nombre': ['Laura', 'Ana', 'María', 'Rosa'],
        'Edad': [25, 30, 35, 40],
        'Ciudad': ['Medellín', 'Bogotá', 'Barranquilla', 'Bucaramanga']}

df = pd.DataFrame(data)
df
```

```python
# Crear un DataFrame a partir de un diccionario de series
# Tomará las claves como columnas
d = {"one": pd.Series([1.0, 2.0, 3.0], index=["a", "b", "c"]),
    "two": pd.Series([1.0, 2.0, 3.0, 4.0], index=["a", "b", "c", "d"])}
print(d, '\n')

df = pd.DataFrame(d)
df
```

```python
# ¿Qué pasa si al crear el DataFrame anterior le pasamos índices o columnas?
```

```python
# Índices
df = pd.DataFrame(d, index=['d', 'a'])
df
```

```python
# Columnas
df = pd.DataFrame(d, columns=['primera', 'segunda'])
df
```

```python
# Crear un DataFrame a partir de un diccionario de arrays de numpy de 1 dimensión
# Tomará las claves como columnas
dicc_arrays = {"columna 1": [1.0, 2.0, 3.0, 4.0], "columna dos": [4.0, 3.0, 2.0, 1.0]}

pd.DataFrame(dicc_arrays)
```

```python
# Índices
df = pd.DataFrame(dicc_arrays, index=['a', 'b', 'c', 'd'])
df
```

```python
# Columnas
df = pd.DataFrame(dicc_arrays, columns=['primera', 'segunda'])
df
```

## Operaciones básicas en Pandas

A continuación, veremos algunas operaciones básicas que se pueden realizar en Pandas.

```python
# Crear un DataFrame de ejemplo
data = {'Nombre': ['Amara', 'Bruno', 'Carlos', 'David', 'Ernesto', 'Fernanda'],
        'Edad': [25, 30, 35, 40, 21, 15],
        'Puntuación': [90, 85, 88, 92, 70, 94]}

df = pd.DataFrame(data)
df
```

### Acceder por índices

Si tu DataFrame se llama df:

**df.iloc[filas, columnas]**

```python
# Acceder a la primera fila
fila_0 = df.iloc[0]

# Acceder a la segunda columna
columna_1 = df.iloc[:, 1]

# Acceder a una celda específica (por ejemplo, la celda en la tercera fila y segunda columna)
celda = df.iloc[2, 1]

# Acceder a un subconjunto del DataFrame (por ejemplo, las primeras dos filas y las primeras dos columnas)
subconjunto = df.iloc[:2, :2]

# Mostrar los resultados
print("Primera fila:\n", fila_0, '\n')
print("\nSegunda columna:\n", columna_1, '\n')
print("\nCelda específica:\n", celda, '\n')
print("\nSubconjunto:\n", subconjunto)
```

### Visualización de los primeros datos

Si tu DataFrame se llama df:

```python
# Por defecto df.head() muestra las 5 primeras filas del DataFrame
df.head()
```

### Columnas

**Visualizar una columna**

```python
# La sintaxis es la misma que si "df" fuera un diccionario y
# quisieramos imprimir el valor asociado a la clave "Nombre"

df['Nombre']
```

```python
# Podemos hacer splice

df['Nombre'][:3]
```

**Agregar una columna**

```python
# También es como si df fuera un diccionario
df['Estado civil'] = ['Soltera', 'Casado', 'Viudo', 'Casado', 'Soltero', 'Soltera']
df
```

```python
# Podemos hacer operaciones entre columnas
df['#'] = df['Edad'] + df['Puntuación']
df
```

```python
# Si son strings el símbolo + las concatena
df['#'] = df['Nombre'] + df['Estado civil']
df
```

```python
# También podemos hacer operaciones que devuelven booleanos
df['Aprobaron'] = df['Puntuación'] > 86
df
```

```python
# Si le pasamos un solo valor lo pondrá en todas las filas
df["NuevaColumna"] = "Holaaaa"
df
```

**Borrar una columna**

```python
# También es como si fuera un diccionario
del df['#']
del df['NuevaColumna']
df
```

### Filtrado de datos

Si tu DataFrame se llama df:

**df.loc[filtro que quieres aplicar]**
* Un filtro nos permitirá mostrar solo algunas filas del DataFrame
* Un filtro es una condición que se declara sobre alguna de las columnas

```python
# Filtrar por edad mayor a 30
df.loc[df['Edad'] > 30]
```

```python
# Filtrar por estado civil igual a 'Soltera'
df.loc[df['Estado civil'] == 'Soltera']
```

```python
# Filtrar por aquellos que aprobaron
df.loc[df['Aprobaron'] == True]
```

```python
# Filtrar por aquellos que son solteros y tienen una puntuación menor a 80
df.loc[(df['Estado civil'] == 'Soltero') & (df['Puntuación'] < 80)]
```

### Operaciones sobre los valores de una columna

```python
# Calcular la media de las edades
df['Edad'].mean()
```

```python
# Traer la edad mínima
df['Edad'].min()
```

```python
# .describe() genera un resumen estadístico que incluye la media, desviación estándar, mínimo, máximo y cuartiles
df['Edad'].describe()
```

```python
# Le suma 5 a todas las edades
df['Edad'].add(5)
```

```python
# Notemos que la operación anterior no afectó el dataframe original
df['Edad']
```

```python
# Le sumamos las edades al puntaje
df['Puntuación'].add(df['Edad'])
```

```python
# Convertimos la columna Nombre en una lista
df['Nombre'].tolist()
```

## Filas

**Agregar una nueva fila**

Debemos crear un DataFrame de una sola fila y luego concatenarlo con el DataFrame original.

```python
# Escribimos una fila nueva
persona_nueva = {'Nombre': 'Germán', 'Edad': 36, 'Puntuación': 87, 'Aprobaron':True}
persona_nueva_df = pd.DataFrame(persona_nueva, index=[0])
persona_nueva_df
```

```python
# Pegamos la fila nueva al DataFrame de bebidas
# Ponemos ignore_index=True para que resetee los índices después de pegarlos
df_2 = pd.concat([df, persona_nueva_df], ignore_index=True)
df_2
```

```python
# ¿Qué pasa si no ponemos ignore_index=True?
df_3 = pd.concat([df, persona_nueva_df])
df_3
```

```python
# Reseteemos los índices de df_3
df_3 = df_3.reset_index()
df_3
```

```python
# Ahora borremos la fila de indices vieja
del df_3['index']
df_3
```
